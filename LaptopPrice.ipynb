```python
df=df.drop("Cpu",axis=1)
```

```python
df["CPU Frequency"]=df["CPU Frequency"].str[:-3]
```

```python
df["Ram"]=df["Ram"].astype("int")
df["CPU Frequency"]=df["CPU Frequency"].astype("float")
df["ScreenWidth"]=df["ScreenWidth"].astype("int")
df["ScreenHeight"]=df["ScreenHeight"].astype("int")
```

```python
df["Memory Amounts"]=df.Memory.str.split(" ").apply(lambda x:x[0])
df["Memory Type"]=df.Memory.str.split(" ").apply(lambda x:x[1:])
```

```python
df=df.drop("Memory",axis=1)
```

```python
df
```

```python
def memory_to_mb(value):
    if "GB" in value :
        return float(value[:value.find("GB")])*1000
    elif "TB" in value :
        return float(value[:value.find("TB")])*1000000
```

```python
df["Memory Amounts"]=df["Memory Amounts"].apply(memory_to_mb)
```

```python
df
```

```python
df
```

```python
df["Weight"]=df["Weight"].str[:-2]
```

```python
plt.figure(figsize=(18, 15))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap="YlGnBu")
```

```python
# Select numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Compute absolute correlations with target column
target_correlations = correlation_matrix['Price_euros'].apply(abs).sort_values()

print(target_correlations)
```

```python
selected_features = target_correlations[-21:].index
```

```python
selected_features = list(selected_features)
```

```python
selected_features
```

```python
limited_df = df[selected_features]
```

```python
limited_df
```

```python
plt.figure(figsize=(18, 15))
sns.heatmap(limited_df.corr(), annot=True, cmap="YlGnBu")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X, y = limited_df.drop("Price_euros", axis=1), limited_df["Price_euros"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
forest = RandomForestRegressor()

forest.fit(X_train_scaled, y_train)
```

```python
forest.score(X_test_scaled, y_test)
```

```python
y_pred = forest.predict(X_test_scaled)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(y_pred, y_test)
plt.plot(range(0, 6000), range(0, 6000), c="red")
```

```python
X_new_scaled = scaler.transform([X_test.iloc[0]])
```
